# visualization

## accuracy

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Base Solver": ["Llama-3.2-3B", "Qwen-2.5-3B", "Gemma-3-4B"],
    "No-Rewrite": [0.380, 0.500, 0.700],
    "Untrained Rewriter": [0.420, 0.360, 0.560],
    "Trained Rewriter": [0.440, 0.320, 0.600],
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(8, 5))

bar_width = 0.25
x = range(len(df))

ax.bar([i - bar_width for i in x], df["No-Rewrite"], width=bar_width, label="No-Rewrite")
ax.bar(x, df["Untrained Rewriter"], width=bar_width, label="Untrained Rewriter")
ax.bar([i + bar_width for i in x], df["Trained Rewriter"], width=bar_width, label="Trained Rewriter")

ax.set_xticks(x)
ax.set_xticklabels(df["Base Solver"])
ax.set_ylabel("Accuracy (Exact Match)")
ax.set_title("Accuracy on GSM8K mini test set (n=50)")
ax.grid(True)
ax.legend()

plt.tight_layout()
plt.show()
```

## token

```python
data_tokens = {
    "Base Solver": ["Llama-3.2-3B", "Qwen-2.5-3B", "Gemma-3-4B"],
    "No-Rewrite": [3055, 3120, 3141],
    "Untrained Rewriter": [2141, 2214, 2273],
    "Trained Rewriter": [2168, 2245, 2304],
}

df_tokens = pd.DataFrame(data_tokens)

fig, ax = plt.subplots(figsize=(8, 5))

bar_width = 0.25
x = range(len(df_tokens))

ax.bar([i - bar_width for i in x], df_tokens["No-Rewrite"], width=bar_width, label="No-Rewrite")
ax.bar(x, df_tokens["Untrained Rewriter"], width=bar_width, label="Untrained Rewriter")
ax.bar([i + bar_width for i in x], df_tokens["Trained Rewriter"], width=bar_width, label="Trained Rewriter")

ax.set_xticks(x)
ax.set_xticklabels(df_tokens["Base Solver"])
ax.set_ylabel("Total prompt-time tokens (lower is better)")
ax.set_title("Prompt-time Tokens by Base Solver")
ax.grid(True)
ax.legend()

plt.tight_layout()
plt.show()
```
